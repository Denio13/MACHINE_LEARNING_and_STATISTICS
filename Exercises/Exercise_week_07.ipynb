## 1. Parameters <a class="anchor" id="Parameters"></a>
***
Let's learn about Parameters. I am going to use `deg` parameter from the ``numpy.polyfit``$^{[1]}$ with different values.

```python
# Import matplotlib [2]
import matplotlib.pyplot as plt

# import the numpy package [3]
import numpy as np
```

```python
# Change plot size
plt.rcParams['figure.figsize'] = [14, 7]

# Background colours and styles for plots
plt.style.use("ggplot")
```

```python
# A range of x values.
x = np.linspace(0.0, 15.0, 30)
x
```

```python
# A range of y values based on x.
y = 6.0 * x - 2.5 + np.random.normal(0.0, 2.0, len(x))
y
```

```python
# Plot x vs y.
plt.plot(x, y, 'k.');
```

```python
# Use x and y to try to find the original parameters 
# Degree of the fitting polynomial is 1
params = np.polyfit(x, y, 1)
params
```

```python
# Plot x vs y.
plt.plot(x, y, 'k.')

# Plot x vs parameters on x.
plt.plot(x, params[0] * x + params[1], 'r-');
```

```python
# Use x and y to try to find the original parameters
# Degree of the fitting polynomial is 2
params = np.polyfit(x, y, 2)
params
```

```python
# Plot x vs y.
plt.plot(x, y, 'k.')

xtmp = np.linspace(-100.0, 100.0, 1000)

# Plot x vs parameters on x.
plt.plot(xtmp, params[0] * xtmp**2 + params[1] * xtmp + params[2], 'r-');
```

```python
# Use x and y to try to find the original parameters
# Degree of the fitting polynomial is 3
params = np.polyfit(x, y, 3)
params
```

```python
# Plot x vs y.
plt.plot(x, y, 'k.')

# Plot x vs parameters on x.
plt.plot(x, np.polyval(params, x), 'r-');
```

```python
# Use x and y to try to find the original parameters
# Degree of the fitting polynomial is 100
params = np.polyfit(x, y, 100)
params
```

```python
# Plot x vs y.
plt.plot(x, y, 'k.')

# Plot x vs parameters on x.
plt.plot(x, np.polyval(params, x), 'r-');
```

We can notice that the result changes with the parameters even if the input data doesn't change.
